## Split the data into features label:x and target label: y

```python
features=df.iloc[:,0:7]
target=df.iloc[:,7]
```

```python
features[:5]
```

```python
target[:5]
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaling=StandardScaler()
scaled_data=scaling.fit_transform(features)
features=pd.DataFrame(scaled_data,columns=features.columns)
features.head()
```

##  Implement DBSCAN

```python
from sklearn.cluster import DBSCAN
```

## Evaluate the model 
- silhouette score
- Adjusted rand score
- Normalized nutual info score

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import normalized_mutual_info_score
```

## Split the features 
- A and P 
- WK and LK 
- C and LKG 
- A_Coef and LKG

## Features:  A and P 

```python
data=features[['A','P']]
data.head()
```

```python
dbscan_cluster=DBSCAN(eps=0.2,min_samples=4,metric='euclidean',algorithm='auto')
dbscan_cluster.fit(data)

```

```python
dbscan_cluster.labels_
```

```python
total_cluster=len(np.unique(dbscan_cluster.labels_))
print('Total Cluster in DBSCAN: ',total_cluster)
x=data['A']
y=data['P']
plt.figure(figsize=(5,5))
plt.scatter(x,y,c=dbscan_cluster.labels_)
plt.xlabel('A')
plt.ylabel('P')
plt.title('A vs P')
```

```python
score_silhouette=silhouette_score(data,dbscan_cluster.labels_)
score_ami=adjusted_rand_score(dbscan_cluster.labels_,target)
score_nmi=normalized_mutual_info_score(dbscan_cluster.labels_,target)

print('Silhouette score: ',score_silhouette)
print('AMI score: ',score_ami)
print('NMI score: ',score_nmi)
```
